```python
from tkinter import *
from tkinter import messagebox
```

## URL shortener GUI

```python
from pyshorteners import *
from tkinter import *
```

```python
url = Shortener()
master = Tk()
```

```python
def shorten():
    ##get the copied url from long_url
    ##convert that to short url
    ##print that in the short_url
    long = long_url.get()
    shortened = url.tinyurl.short(long)
    short_url.insert(0,shortened)
```

### 3 Labels, 2 entry fields ,1-Button

```python
label1 = Label(master,text = "URL Shortener",font = ("Helvetica",16,"bold"),padx=20,pady=10).grid(row=0,column=0,columnspan=2)
label2 = Label(master,text = "Copy the URL here",font=("Helvetica",11,"bold italic")).grid(row=1,column=0)
long_url = Entry(master)
label2 = Label(master,text="Get the tinyURL here",font=("Helvetica",11,"bold italic")).grid(row=2,column=0)
short_url = Entry(master)
Generate = Button(master,text="Generate",font=("Arial",12,"bold"),command = shorten).grid(row=3,column = 0)
```

```python
long_url.grid(row=1,column=1,ipadx=35,ipady=10)
short_url.grid(row=2,column=1,ipadx=35,ipady=10)
```

```python
master.mainloop()
```
